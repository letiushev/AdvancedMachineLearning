```python
import csv
with open('log.csv', newline='') as f:
    reader = csv.reader(f)
    data = list(reader)
rewards = [0] * len(data)
epsilon = [0] * len(data)
for i in range(len(data)):
    rewards[i] = data[i][0]
    epsilon[i] = data[i][1]
```

```python
import pandas as pd
import matplotlib.pyplot as plt

iterations = [0] * 25
for i in range(25):
    iterations[i] = i+1

df = pd.DataFrame.from_dict({
    'Iteration': iterations,
    'Reward': rewards,
})
```

Evaluation for training

```python
df = pd.read_csv('D:\Programming\AML\log copy.csv')

# Draw Plot
plt.figure(figsize=(16,10), dpi= 80)
plt.plot('iteration', 'reward', data=df, color='tab:red')

# Decoration
plt.ylim(0, 600)
xtick_location = df.index.tolist()[::12]
#xtick_labels = [x[-4:] for x in df.iteration.tolist()[::12]]
#plt.xticks(ticks=xtick_location, labels=xtick_labels, rotation=0, fontsize=12, horizontalalignment='center', alpha=.7)
plt.xticks(ticks=xtick_location, rotation=0, fontsize=12, horizontalalignment='center', alpha=.7)
plt.yticks(fontsize=12, alpha=.7)
plt.title("Rewards", fontsize=22)
plt.grid(axis='both', alpha=.3)

# Remove borders
plt.gca().spines["top"].set_alpha(0.0)    
plt.gca().spines["bottom"].set_alpha(0.3)
plt.gca().spines["right"].set_alpha(0.0)    
plt.gca().spines["left"].set_alpha(0.3)   
plt.show()
```
